```python
import utils
import models
import os 
import cv2
import random
import matplotlib.pyplot as plt 

import keras.backend as K
import keras.losses as KL
from keras.preprocessing.image import ImageDataGenerator
```

```python
IMAGE_DIR = 'D:\\projects\\datasets\\bbox_regression\\images'
ANNOTATION_DIR = 'D:\\projects\\datasets\\bbox_regression\\annotations'

train_datagen = ImageDataGenerator(rescale=1./255)
train_generator = train_datagen.flow_from_directory(IMAGE_DIR, target_size=(224, 224), batch_size=4, class_mode='categorical')
```

```python
model = models.build_model()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
history = model.fit_generator(train_generator, steps_per_epoch=10, epochs=30)
```

```python
x_data, y_data = utils.load_dataset(IMAGE_DIR, ANNOTATION_DIR, dsize=(224, 224))
x_data, y_data = utils.shuffle_data(x_data, y_data)
```

```python
model.evaluate(x_data, y_data[:,4:])
```
